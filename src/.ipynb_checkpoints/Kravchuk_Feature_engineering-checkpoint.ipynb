```python
ds.shape
```

```python
for col in ds.columns:
    if ds[col].isnull().values.any():
        print("Missing data in ", col, ds[col].isnull().sum())
```

```python
# створимо функцію для заповнення відсутніх значень:
def impute_na(df, variable, value):
    return df[variable].fillna(value)
```

Ця функція приймає датасет, назву стовпця, в якому необхідно замінити відсутні значення (variable) та значення, якими ми заміняємо відсутні значення (value). Функція повертає оновлений стовпець датасету з заміненими відсутніми значеннями.

Вибір між медіаною, середнім та модою для заміни відсутніх значень залежить від розподілу змінної та наявності викидів. Зазвичай середнє значення використовують для змінних з нормальним розподілом та без значущих викидів, медіану - для змінних зі зміщеним розподілом або коли велика кількість викидів, та моду - для категоріальних змінних.

```python
# Знаходження медіани для стовпця
median_arrival_delay = ds['Arrival Delay in Minutes'].median()
```

```python
# Заміна пропущених значень на медіану
ds['Arrival Delay in Minutes'] = impute_na(ds, 'Arrival Delay in Minutes', median_arrival_delay)
```

```python
for col in ds.columns:
    if ds[col].isnull().values.any():
        print("Missing data in ", col, ds[col].isnull().sum())
    else:
        print("There aren't missing data")
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

ds["Gender"] = le.fit_transform(ds["Gender"])
ds["Customer Type"] = le.fit_transform(ds["Customer Type"])
ds["Type of Travel"] = le.fit_transform(ds["Type of Travel"])
ds["Class"] = le.fit_transform(ds["Class"])
ds["satisfaction"] = le.fit_transform(ds["satisfaction"])
```

## Опрацювання аномалій


Я використаю метод "Local Outlier Factor" для виявлення та видалення викидів.

```python
from sklearn.neighbors import LocalOutlierFactor

numerical_columns = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']

# Ініціалізація моделі LocalOutlierFactor
lof = LocalOutlierFactor(contamination=0.05)  # Встановіть contamination залежно від розміру викидів, які ви очікуєте

# Обчислення оцінки аномалій (викидів) для кожного прикладу в датасеті
outlier_scores = lof.fit_predict(ds[numerical_columns])

# Відображення кількості викидів
outlier_count = len(ds[outlier_scores == -1])
print("Number of outliers:", outlier_count)

# Видалення рядків, що містять викиди
ds = ds[outlier_scores != -1]

# Відображення оновленого датасету без викидів
print(f"Dataset without outliers:{len(ds)}")
```

## Функція масштабування


Ми використовуємо масштабування ознак (Feature Scaling), щоб стандартизувати незалежні ознаки в фіксованому діапазоні, щоб кожна ознака мала приблизно однаковий внесок у модель. З огляду на розподіл моїх даних, ми не можемо використовувати Standard Scaler. Ми будемо використовувати MinMax Scaler, який перетворює діапазон даних на (0,1).

```python
from sklearn.preprocessing import MinMaxScaler


numerics=['Departure Delay in Minutes', 'Arrival Delay in Minutes','Flight Distance',"Age"]

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(ds[numerics])

# Перетворені дані зберігаються у вигляді numpy-масиву, створимо DataFrame для зручності
scaled_df = pd.DataFrame(scaled_data, columns=numerics)

ds[numerics] = scaled_df[numerics]
```
